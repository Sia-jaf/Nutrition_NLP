## Name and Score the Clusters

```python
# give categories (clusters) names instead of digits
# also give a score for cluster heterogeneity
# shuffle the clusters so the scorer doesn't know which is which
clusters = [1 for _ in range(30)]
print(clusters)
clusters.extend([2 for _ in range(30)])
print(clusters)
random.shuffle(clusters)
print(clusters)
print(len(clusters))
```

```python
scores = {'kmm30': [],
          'k30': []}
maps = {}
maps['kmm30'] = {k: '' for k in range(30)}
maps['k30'] = {k: '' for k in range(30)}
```

```python
k30s = 0
kmm30s = 0
for k in range(60):
    cat = 'more'
    while cat == 'more':
        if clusters[k] == 1:
            if maps['kmm30'][kmm30s] != '':
                kmm30s += 1
                cat = 'continue'
                continue
            # assess a kmm30 cluster
            samp = ('\n'.join(df.loc[df['kmm30'] == kmm30s, 'description']
                    .sample(n = 25,replace=True).to_list()))
            print('\n')
            print(samp)
            cat = input('type category, "break", or "more": ')
            if cat == 'break':
                break
            maps['kmm30'][kmm30s] = cat
            sco = input("how many foods don't fit?")
            scores['kmm30'].append(sco)
            kmm30s += 1
        if clusters[k] == 2:
            if maps['k30'][k30s] != '':
                k30s += 1
                cat = 'continue'
                continue
            # assess a k30 cluster
            samp = ('\n'.join(df.loc[df['k30'] == k30s, 'description']
                    .sample(n = 25).to_list()))
            print('\n')
            print(samp)
            cat = input('type category, "break", or "more": ')
            if cat == 'break':
                break
            maps['k30'][k30s] = cat
            sco = input("how many foods don't fit?")
            scores['k30'].append(sco)
            k30s += 1

    if cat == 'break':
        break


```
