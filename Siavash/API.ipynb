```python
# Source - https://stackoverflow.com/a/62377768
# Posted by Nicolas M., modified by community. See post 'Timeline' for change history
# Retrieved 2026-02-16, License - CC BY-SA 4.0

foodName = "Cheddar cheese"

def call_API(foodName, apiKey):
    url = f'https://api.nal.usda.gov/fdc/v1/foods/search?api_key={apiKey}&query={foodName}'
    r = req.get(url)
    print(r.status_code)  # 200
    return r.json()


def call_API_2(foodName, apiKey):
    data = {"query" : foodName}
    url = f'https://api.nal.usda.gov/fdc/v1/foods/search?api_key={apiKey}'
    r = req.post(url, json=data)
    print(r.status_code)  # 200
    return r.json()

ans = call_API_2(foodName, apiKey)
ans.keys()
```

```python
df = pd.json_normalize(ans, 'foods')
dff = df['foodNutrients'][0][0]
print(dff.keys())
list = dff.values()
print(list)
```
